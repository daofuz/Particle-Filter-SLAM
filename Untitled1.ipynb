```python
import numpy as np
import matplotlib.pyplot as plt
gdmap = np.loadtxt("data.txt").reshape(1500, 1400)
possv = np.loadtxt("data_pos.txt").reshape(1161, 3)
vmin,vmax = 0,2
plt.figure(figsize=(15,15))
grid_tmp = 1-(gdmap<0)
grid_tmp[grid_tmp==vmin] = vmax
plt.imshow(grid_tmp.T, cmap='ocean', vmin=vmin, vmax=vmax,origin="lower")
plt.plot(possv[:1160, 0]+100, possv[:1160, 1]+1200, color="red",label="trajectory", linewidth=3)
plt.axis("off")
plt.show()
```
